## Imports

```python

import numpy as np
import matplotlib.pyplot as plt
```

## Linear convection without CFL

```python
def linearconv_withoutCFL(u, nx, nt, dx):
    
    dt =0.025  
    c = 1. 
    for n in range(nt):  
        un=u.copy()

        for i in range(1, nx):
            u[i] = un[i] - c * dt / dx * (un[i] - un[i-1])
    
    return 0
```

## Linear convection with CFL

```python
def linearconv_withCFL(u, nx, nt, dx):
  
    CFL = 0.8
    c = 1.
    dt = CFL*dx/np.fabs(c) 
    for n in range(nt):  #Time ITERATION
        un=u.copy()
        for i in range(1, nx):
            u[i] = un[i] - c * dt / dx * (un[i] - un[i-1])  
    return 0
        
```

## Inputs

```python

nx = 85    #change to 83, 85, 91, ... what do you remark ?

nt = 20  
dx = 2 / (nx - 1)
grid = np.linspace(0,2,nx) 
u0 = np.ones(nx)      
u0[int(.5/dx):int(1 / dx + 1)] = 2  
```

## Call Funtions

### Function without CFL

```python
u = u0.copy()
linearconv_withoutCFL(u, nx, nt, dx)
us=u.copy()
plt.plot(grid,u0);
plt.plot(grid,us);
plt.ylim([1.,2.2])
plt.xlabel('X')
plt.ylabel('Velocity')    
plt.title('1D Linear Convection')  
```

### Function with CFL

```python
u1= u0.copy()
linearconv_withCFL(u1, nx, nt, dx)
plt.plot(grid,u0);
plt.plot(grid,u1);
plt.ylim([1.,2.2])
plt.xlabel('X')
plt.ylabel('Velocity')    
plt.title('1D Linear Convection') 
```
